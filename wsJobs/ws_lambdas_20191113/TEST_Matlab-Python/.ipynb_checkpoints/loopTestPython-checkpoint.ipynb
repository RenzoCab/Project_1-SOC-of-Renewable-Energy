```python
comm = numpy.asarray([12,0])
numpy.savetxt("loop.csv", comm, delimiter=",")
while 1 == 1:
    time.sleep(3)
    rad = numpy.loadtxt("loop.csv", delimiter =', ')
    if float(rad[-1]) == 3:
        print('Matlab sent a 3!')
    if random.random() < 0.2:
        comm = numpy.asarray([12,2])
    else:
        comm = numpy.asarray([12,0])
    numpy.savetxt("loop.csv", comm, delimiter=",")
```
